```python
def bar_chart_words(words, top=10, messages_type="", color="#1f77b4"):
    top_spam = np.array(sorted(words.items(), key=lambda x: -x[1]))[:top]
    top_words = top_spam[ : :-1, 0]
    top_words_count = [int(i) for i in top_spam[ : :-1, 1]]
    # aesthetics
    if messages_type:
        messages_type = messages_type + " " 
    plt.title(f"Top {top} most common words in {messages_type} comments")
    plt.xlabel(f"Number of words")
    plt.barh(top_words, top_words_count, color=color)
    plt.show()
```

#### همانطور که دیده میشود طول پیام های تایید شده بیشتر از ریجکت شده ها است 

```python
stop_words = hazm.stopwords_list()

def RepresentsInt(s):
    try: 
        int(s)
        return True
    except ValueError:
        return False


def process_comment(comment):
    words = re.split(r'\W+',comment)
    words = [word for word in words if len(word) > 2]
#     words = [word for word in words if word not in stop_words]
    words = [word for word in words if not RepresentsInt(word)]
    return words
```

```python
data.comment_title.apply(process_comment).head()
# print(data['comment'])
```

```python
def count_processed_words(data):
    counter = collections.OrderedDict()
    for comment in data:
        words = process_comment(comment)
        for word in set(words):
            if (word in counter):
                counter[word] += 1
            else:
                counter[word] = 1
    return counter
```

```python
verified_comments = set(data[data.verification_status == 'verified']['comment_title'])
verified_words = count_processed_words(verified_comments)

rejected_comments = set(data[data.verification_status == 'rejected']['comment_title'])
rejected_words = count_processed_words(rejected_comments)
```

```python
bar_chart_words(verified_words, top=15 , messages_type="verifed", color= "green" )
```

```python
bar_chart_words(rejected_words, top=15 , messages_type="rejected" , color= "red" )
```

# Classify

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
from sklearn.model_selection import train_test_split

comment_train, comment_test, verification_status_train, verification_status_test = \
train_test_split(data["comment"], data["verified"], test_size=0.2, random_state=5)

comment_title_train, comment_title_test, verification_status_train, verification_status_test = \
train_test_split(data["comment_title"], data["verified"], test_size=0.2, random_state=5)
```

## Naive Bayesian

```python
pipe_MNB = Pipeline([ ('bow'  , CountVectorizer(analyzer = process_comment) ),
                   ('tfidf'   , TfidfTransformer()),
                   ('MNB' , MultinomialNB()),
                    ])
```

### Comment 

```python
pipe_MNB.fit(X=comment_train, y=verification_status_train)

pred_test_MNB = pipe_MNB.predict(comment_test)
acc_MNB = accuracy_score(verification_status_test, pred_test_MNB)
print(acc_MNB)
print(pipe_MNB.score(comment_test, verification_status_test))
```

### Comment + Title

```python
pipe_MNB.fit(X=comment_title_train, y=verification_status_train)

pred_test_MNB = pipe_MNB.predict(comment_title_test)
acc_MNB = accuracy_score(verification_status_test, pred_test_MNB)
print(acc_MNB)
print(pipe_MNB.score(comment_title_test, verification_status_test))
```

## SVC

## Comment + Title 

```python
pipe_SVC = Pipeline([ ('bow'  , CountVectorizer(analyzer = process_comment) ),
                   ('tfidf'   , TfidfTransformer()),
                   ('clf_SVC' , SVC(gamma='auto', C=1000)),
                    ])

parameters_SVC = dict(tfidf=[None, TfidfTransformer()],
                      clf_SVC__C=[500, 1000,1500]
                      )
grid_SVC = GridSearchCV(pipe_SVC, parameters_SVC, 
                        cv=5, n_jobs=-1, verbose=1)

grid_SVC.fit(X=comment_title_train, y=verification_status_train)
```

```python
grid_SVC.best_score_
```

```python
pred_test_grid_SVC = grid_SVC.predict(comment_title_test)
acc_SVC = accuracy_score(verification_status_test, pred_test_grid_SVC)
print(acc_SVC)
print(grid_SVC.score(comment_title_test, verification_status_test))
```

```python
count = 0
for i in range(len(verification_status_test)):
    if verification_status_test.iloc[i] != pred_test_grid_SVC[i]:
        count += 1
print('Total number of test cases', len(verification_status_test))
print('Number of wrong of predictions', count)
```

```python
from sklearn.metrics import classification_report
print(classification_report(pred_test_grid_SVC, verification_status_test))
```
